```
import numpy as np
import pandas as pd
import plotly.express as px
import plotly.figure_factory as ff
```

```
dane = pd.read_csv('/content/red_wine.csv')
dane
```

```
dane.describe()
```

```
zq = dane['quality']
zbior_quality = np.array(zq)
zbior_quality
```

```
ff.create_distplot([zbior_quality],['Quality'])
#XDDD
```

```
najlepsze_wino = dane.sort_values('quality',ascending = False)
najlepsze_wino = najlepsze_wino.head(1)
najlepsze_wino
#bez sensu powyższe. tą samą ocene moze miec kilka win. -> zbadać grupę.
```

```
najlepsze_wina = dane[dane['quality'] == 8]
najlepsze_wina
```

```
najlepsze_wina.describe()
#średnie statystyki w wierszu "mean".
```

```
najgorsze_wina = dane.sort_values('quality',ascending = False)
najgorsze_wina = najgorsze_wina[najgorsze_wina['quality'] == 3]
najgorsze_wina
```

```
najgorsze_wina.describe()
#średnie statystyki najgorszych win w kolumnie "mean".
```

```
#czy mocniejsze wino ma lepsze oceny niż słabsze?

px.box(dane,
       x = 'quality',
       y = 'alcohol',
       title = 'ocena ze względu na moc wina.')

#na ponizszym wykresie widać, że alkohol z "quality" = 8 ma więcej wyników w wysokoprocentowych winach -> mediana 12.15%, grupa niskoprocentowych jest nieliczna w ocenie 8.
```

```
#czy mocniejsze wino ma lepsze oceny niż słabsze?

px.scatter(dane,
           x = 'quality',
           y= 'alcohol',
           )

#widzimy, że wina powyżej 13%% praktycznie nie występują w jakości poniżej 6.
```

```
korelacja = dane.corr()
korelacja
```

```
'''
korelacje..

density -> fixed.acidity = 0.66
citric.acid -> fixed.acidity = 0.67
total.sulfur.dioxide -> free.sulfur.dioxide = 0.66 

dodatkowo - > jakość zależy od poziomu alkoholu w 47%

'''
px.imshow(korelacja)
```

```
korelacja['quality'].sort_values()
```
